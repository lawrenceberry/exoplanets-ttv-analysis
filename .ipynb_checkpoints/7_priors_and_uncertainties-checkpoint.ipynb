### Using priors and parameterizing uncertainty to deal better with outliers

In the ideal case where you can directly observe your model parameters T0, P0 and dP/dE, a posterior distribution over the mean and variance of these parameters takes the form of the Wishart distribution.

Let's try and tackle this as a Bayesian multivariate regression problem: https://en.wikipedia.org/wiki/Bayesian_multivariate_linear_regression

Or using MCMC, and a uniform prior on measurement uncertainty greater than (but not smaller than) the given measurement error:

##### WASP-12b

```python
df = pd.read_csv("./data/ORIGINALS/ETD_WASP-12_og.txt", sep="\t", names=["index", "epoch", "transit time", "error", "data quality", "observer"])
df["transit time"] = df["transit time"] - df["transit time"].quantile(0.5)
df["epoch"] = df["epoch"] - df["epoch"].quantile(0.5)

observations = df[["epoch", "transit time"]]
observations = observations.set_index(df.epoch.astype(int))
observations = observations["transit time"]
error = df["error"].values
```

```python
def model(E, T0, P0, dP_dE):
    return T0 + P0*E + 0.5*(dP_dE/1000/60/60/24)*E**2

def log_likelihood(theta):
    T0, P0, dP_dE, *sd = theta
    
    N = len(observations.values)
    E = observations.index.values
    mu = model(E, T0, P0, dP_dE)
    
    diff = observations.values - mu
    log_likelihood = -(0.5*N*np.log(2*np.pi)) -0.5*(np.sum(np.power(sd, 2))) -(0.5*diff@((1/np.power(sd, 2))*diff))

    return log_likelihood

def neg_ML_log_likelihood(theta):
    T0, P0, dP_dE = theta
    
    N = len(observations.values)
    E = observations.index.values
    mu = model(E, T0, P0, dP_dE)
    
    diff = observations.values - mu
    log_likelihood = -(0.5*N*np.log(2*np.pi)) -0.5*(np.sum(np.power(error, 2))) -(0.5*diff@((1/np.power(error, 2))*diff))

    return -log_likelihood

def log_prior_sd(sd):
    if np.any(np.array(sd) < error):
        return -np.inf
    return 0.0  # np.sum(np.log(-2*(sd-error)+(2/FIVE_YEARS)))

def log_prior_p0(P0):
    if P0 < 0:
        return -np.inf  # can't have negative periods
    return 0.0

def log_prior(theta):
    T0, P0, dP_dE, *sd = theta
    return log_prior_p0(P0) + log_prior_sd(sd)

def log_probability(theta):
    return log_prior(theta) + log_likelihood(theta)
```

```python
# first do a maximum likelihood fit to find good initial start parameters
x0 = (0.0, 1.0, 0.0)
res = sp.optimize.minimize(neg_ML_log_likelihood, x0, method="BFGS", tol=1e-6, options={'disp': True, "maxiter": 1e4})
res.x
```

```python
x0 = np.array([0.0, 1.1, 0.0])
pos = x0 + 1e-1 * np.random.randn(750, 3)
error_pos = error + error * 0.5 * np.abs(np.random.randn(750, len(error)))
pos = np.concatenate((pos, error_pos), axis=1)
assert not np.any(pos[:, 3:] < error)
assert not np.any(pos[:, 1] < 0)
```

```python
nwalkers, ndim = pos.shape
print("nwalkers", nwalkers, "ndim", ndim)
sampler = emcee.EnsembleSampler(
    nwalkers, ndim, log_probability, 
)
sampler.run_mcmc(pos, 10_000, progress=True, skip_initial_state_check=True);
```

```python
samples = sampler.get_chain(discard=500, thin=250, flat=True)
samples.shape
```

```python
fig, axes = plt.subplots(3, figsize=(10, 7), sharex=True)

s = samples[:, [0,1,2]]
labels = ["T0", "P0", "dP_dE"]

for i in range(3):
    ax = axes[i]
    ax.plot(s[:, i], "k", alpha=0.3)
    ax.set_xlim(0, len(s))
    ax.set_ylabel(labels[i])

axes[-1].set_xlabel("Step number")
```

```python
error_samples = samples[:, 3:]
map_errors = np.percentile(error_samples, 50, axis=0)
diff = (map_errors-error)*24*60*60
plt.hist(diff, bins=50)
plt.xlabel("$\Delta\sigma$ [s]")
plt.ylabel("Frequency")
plt.show()
```

```python
means = []
variances = []
for i in tqdm(range(100, len(samples))):
    subslice = samples[:i, 2]
    mean = np.mean(subslice)
    means.append((i, mean))
    variance = np.var(subslice)
    variances.append((i, variance))
means = np.array(means)
variances = np.array(variances)

plt.plot(means[:,0], means[:,1])
plt.ylabel("Mean dP/dE")
plt.xlabel("Step number")
plt.show()

plt.plot(variances[:,0], variances[:,1])
plt.ylabel("Variance of dP/dE")
plt.xlabel("Step number")
plt.show()
```

This is telling us we can still fit to the data and get a reasonable result without using the measurement uncertainties, but rather inferring measurement uncertainties during the fit itself. We naturally end up with a variance of ~0.0004, much larger that our original variance of ~0.00002, but this is perhaps a more realistic "lower bound" when we are unsure as to the provided uncertainty of the measurements.

Note however that after 25000 the parameters have still not converged, we need to run more iterations.
